```python
# Smart Library Circulation System

MAX_BOOKS_PER_STUDENT = 3
DUE_DAYS = 14
FINE_PER_DAY = 5

books = {}       # book_id -> book details
students = {}    # student_id -> student name
issued_books = {}  # student_id -> list of issued books


def add_book():
    book_id = input("Enter Book ID: ")
    if book_id in books:
        print("Book ID already exists!")
        return

    title = input("Enter Title: ")
    author = input("Enter Author: ")
    total = int(input("Enter Total Copies: "))

    books[book_id] = {
        "title": title,
        "author": author,
        "total": total,
        "available": total
    }
    print("Book added successfully!")


def register_student():
    student_id = input("Enter Student ID: ")
    if student_id in students:
        print("Student already registered!")
        return

    name = input("Enter Student Name: ")
    students[student_id] = name
    issued_books[student_id] = []
    print("Student registered successfully!")


def issue_book():
    student_id = input("Enter Student ID: ")
    book_id = input("Enter Book ID: ")

    if student_id not in students:
        print("Student not found!")
        return

    if book_id not in books:
        print("Book not found!")
        return

    if books[book_id]["available"] == 0:
        print("No copies available!")
        return

    if len(issued_books[student_id]) >= MAX_BOOKS_PER_STUDENT:
        print("Student has reached maximum book limit!")
        return

    for b in issued_books[student_id]:
        if b[0] == book_id:
            print("This book is already issued to the student!")
            return

    days = int(input("Enter number of days issued: "))
    issued_books[student_id].append((book_id, days))
    books[book_id]["available"] -= 1
    print("Book issued successfully!")


def return_book():
    student_id = input("Enter Student ID: ")
    book_id = input("Enter Book ID: ")

    if student_id not in issued_books:
        print("Student not found!")
        return

    for b in issued_books[student_id]:
        if b[0] == book_id:
            days = b[1]
            issued_books[student_id].remove(b)
            books[book_id]["available"] += 1

            if days > DUE_DAYS:
                fine = (days - DUE_DAYS) * FINE_PER_DAY
                print(f"Book returned with fine: ₹{fine}")
            else:
                print("Book returned successfully. No fine.")
            return

    print("This book was not issued to the student!")


def search_books():
    keyword = input("Enter keyword to search: ").lower()
    found = False

    for book in books.values():
        if keyword in book["title"].lower() or keyword in book["author"].lower():
            print(book)
            found = True

    if not found:
        print("No books found!")


def overdue_report():
    print("\nOverdue Report:")
    for student_id, issued in issued_books.items():
        for book_id, days in issued:
            if days > DUE_DAYS:
                fine = (days - DUE_DAYS) * FINE_PER_DAY
                print(f"Student: {students[student_id]}, Book ID: {book_id}, Fine: ₹{fine}")


def menu():
    while True:
        print("\n--- Smart Library System ---")
        print("1. Add Book")
        print("2. Register Student")
        print("3. Issue Book")
        print("4. Return Book")
        print("5. Search Books")
        print("6. Overdue Report")
        print("7. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            add_book()
        elif choice == "2":
            register_student()
        elif choice == "3":
            issue_book()
        elif choice == "4":
            return_book()
        elif choice == "5":
            search_books()
        elif choice == "6":
            overdue_report()
        elif choice == "7":
            print("Exiting system. Goodbye!")
            break
        else:
            print("Invalid choice!")


menu()
```

```python
# Personal Expense Tracker

expenses = []
MONTHLY_BUDGET = 5000  # You can change this value


def add_expense():
    date = input("Enter date (DD-MM): ")
    category = input("Enter category (Food, Travel, etc): ")
    amount = float(input("Enter amount: "))

    expense = {
        "date": date,
        "category": category,
        "amount": amount
    }

    expenses.append(expense)
    print("Expense added successfully!")


def monthly_summary():
    total = 0
    for e in expenses:
        total += e["amount"]

    print("Total Monthly Spending: ₹", total)

    if total > MONTHLY_BUDGET:
        print("⚠ Budget Breach Alert!")


def category_report():
    category_total = {}

    for e in expenses:
        cat = e["category"]
        if cat in category_total:
            category_total[cat] += e["amount"]
        else:
            category_total[cat] = e["amount"]

    print("\nCategory-wise Spending:")
    for cat, amt in category_total.items():
        print(cat, ":", amt)


def highest_expense_day():
    day_total = {}

    for e in expenses:
        day = e["date"]
        if day in day_total:
            day_total[day] += e["amount"]
        else:
            day_total[day] = e["amount"]

    highest_day = ""
    highest_amount = 0

    for day, amt in day_total.items():
        if amt > highest_amount:
            highest_amount = amt
            highest_day = day

    print("Highest Spending Day:", highest_day, "₹", highest_amount)


def dominant_category():
    category_total = {}

    for e in expenses:
        cat = e["category"]
        if cat in category_total:
            category_total[cat] += e["amount"]
        else:
            category_total[cat] = e["amount"]

    top_category = ""
    max_amount = 0

    for cat, amt in category_total.items():
        if amt > max_amount:
            max_amount = amt
            top_category = cat

    print("Dominant Spending Category:", top_category)


def menu():
    while True:
        print("\n--- Personal Expense Tracker ---")
        print("1. Add Expense")
        print("2. Monthly Summary")
        print("3. Category-wise Report")
        print("4. Highest Expense Day")
        print("5. Dominant Category")
        print("6. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            add_expense()
        elif choice == "2":
            monthly_summary()
        elif choice == "3":
            category_report()
        elif choice == "4":
            highest_expense_day()
        elif choice == "5":
            dominant_category()
        elif choice == "6":
            print("Thank you for using Expense Tracker!")
            break
        else:
            print("Invalid choice!")


menu()
```
